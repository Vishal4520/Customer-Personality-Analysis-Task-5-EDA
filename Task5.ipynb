```python
import pandas as pd            # to work with tables (dataframes)
import matplotlib.pyplot as plt # to make graphs
import seaborn as sns           # for prettier graphs
```

```python
df = pd.read_csv(r"E:\DA Project\cleaned_customer_personality_analysis.xlsx.csv") # load file 
df.head() #head() shows the first 5 rows.
```

```python
num_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Numeric columns:", num_cols)
print("Categorical columns:", cat_cols)                 
      
#Makes two lists â€” one for numeric columns, one for text-based columns.


```

```python
# Sample if dataset is large
sns.pairplot(df[num_cols[:3]], diag_kind='kde')  # first 5 numeric cols for clarity
plt.show()
```

```python
plt.figure(figsize=(15,3))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df[num_cols].hist(bins=20, figsize=(14,10))
plt.tight_layout()
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```
